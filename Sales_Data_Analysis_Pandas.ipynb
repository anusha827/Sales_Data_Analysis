```python
all_data_new=all_data[all_data['Order ID'].duplicated(keep=False)]
all_data_new.head(20)
```

```python
all_data_new['Merged']=all_data_new.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
all_data_new.head()
```

```python
# Dropping duplicates

all_data_new=all_data_new[['Order ID','Merged']].drop_duplicates()
all_data_new.head()
```

```python
# counting no.of combinations

from itertools import combinations
from collections import Counter

count=Counter()

for row in all_data_new['Merged']:
    r_list=row.split(',')
    count.update(Counter(combinations(r_list,2)))
    #count.update(Counter(combinations(r_list,3)))  #for three items together
# print(count)
count.most_common(10)
```

```python
for key,value in count.most_common(10):        # more readable than the above
    print(key,value)

```

#### -> iphone and lightning charging cable are the products often sold together.
